```python
dv_historial_contratos_190411["estudios"] = dv_historial_contratos_190411['estudios'].astype('bool')
```

```python
dv_historial_contratos_190411["fecha_inicio"] = dv_historial_contratos_190411['fecha_inicio'].astype('datetime64')
```

```python
dv_historial_contratos_190411["fecha_fin"] = dv_historial_contratos_190411['fecha_fin'].astype('datetime64')
```

```python
dv_historial_contratos_190411 = dv_historial_contratos_190411[dv_historial_contratos_190411['fecha_fin'].notna()]
```

```python
dv_historial_contratos_190411["duracion_contrato_dias"] = dv_historial_contratos_190411['duracion_contrato_dias'].astype('int64')
```

```python
dv_historial_contratos_190411_dni = dv_historial_contratos_190411.groupby('dni').apply(lambda x: x.sort_values(['fecha_inicio','fecha_fin'],ascending=True));
dv_historial_contratos_190411_dni
```
