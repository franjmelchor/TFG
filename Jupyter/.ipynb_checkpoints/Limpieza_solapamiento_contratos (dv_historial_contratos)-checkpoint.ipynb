```python
import pandas as pd
```

```python
dv_historial_contratos_190411 = pd.read_csv('./../datasets_editados/dv_historial_contratos_190411_procesada.csv',sep='|')
```

```python
dv_historial_contratos_190411.head()
```

```python
dv_historial_contratos_190411.columns
```

```python
dv_historial_contratos_190411.dtypes
```

```python
dv_historial_contratos_190411["estudios"] = dv_historial_contratos_190411['estudios'].astype('bool')
```

```python
dv_historial_contratos_190411["fecha_inicio"] = dv_historial_contratos_190411['fecha_inicio'].astype('datetime64')
```

```python
dv_historial_contratos_190411["fecha_fin"] = dv_historial_contratos_190411['fecha_fin'].astype('datetime64')
```

```python
dv_historial_contratos_190411 = dv_historial_contratos_190411[dv_historial_contratos_190411['fecha_fin'].notna()]
```

```python
dv_historial_contratos_190411["duracion_contrato_dias"] = dv_historial_contratos_190411['duracion_contrato_dias'].astype('int64')
```

```python
dv_historial_contratos_190411_dni = dv_historial_contratos_190411.groupby('dni').apply(lambda x: x.sort_values(['fecha_inicio','fecha_fin'],ascending=True));
dv_historial_contratos_190411_dni
```

```python
dv_historial_contratos_190411_dni.set_index('id')
dv_historial_contratos_190411_dni
```

```python
dv_prueba = dv_historial_contratos_190411_dni.loc[dv_historial_contratos_190411_dni['dni'] == '1a6392217b80d649033291334d17b46c6929ff7d55c5ddc859422e7cd73aa96cb8b8fb999bc0958181339e53c6d1c7c69e7f5956dbe8815ac68fdf9a617baeda']
dv_prueba = dv_prueba.loc[:,['id','fecha_inicio','fecha_fin']]
dv_prueba = dv_prueba.iloc[1:8]
```

```python
list = []
for row in dv_prueba.itertuples():
    dni = row.dni
    if dni not in list:
        list.append(dni)
        dv_prueba_dni_x = dv_prueba.loc[dv_prueba.dni['dni'] == dni];
        model = row
        modelFecha_fin = model.fecha_fin
        for row in dv_prueba_dni_x.itertuples():
            if(row.fecha_inicio < model.fecha_fin):
                if(row.fecha_fin > modelFecha_fin):
                    dv_historial_contratos_190411_dni.loc[(dv_historial_contratos_190411_dni.id == model.id),'fecha_fin'] = row.fecha_fin
                    modelFecha_fin = row.fecha_fin
                dv_historial_contratos_190411_dni.drop(dv_historial_contratos_190411_dni[dv_historial_contratos_190411_dni.id == row.id].index, inplace=True)
            else:
                model = row
                modelFecha_fin = row.fecha_fin
                
```

```python
list = []
for row in dv_historial_contratos_190411_dni.itertuples():
    dni = row.dni
    if dni not in list:
        list.append(dni)
        dv_historial_contratos_dni_x = dv_historial_contratos_190411_dni[dv_historial_contratos_190411_dni.dni.eq(dni)];
        model = row
        modelFecha_fin = model.fecha_fin
        for row in dv_historial_contratos_dni_x.itertuples():
            if(row.fecha_inicio < model.fecha_fin):
                if(row.fecha_fin > modelFecha_fin):
                    dv_historial_contratos_190411_dni.loc[(dv_historial_contratos_190411_dni.id == model.id),'fecha_fin'] = row.fecha_fin
                    modelFecha_fin = row.fecha_fin
                dv_historial_contratos_190411_dni.drop(dv_historial_contratos_190411_dni[dv_historial_contratos_190411_dni.id == row.id].index, inplace=True)
            else:
                model = row
                modelFecha_fin = row.fecha_fin
                
```
